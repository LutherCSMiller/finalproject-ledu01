```python
import pandas as pd
import numpy as np
import json
import requests
```

```python
df = pd.read_csv("Directory_Of_Toilets_In_Public_Parks (1).csv")
```

```python
df = df[df['Open Year-Round'] == "Yes"][['Name', 'Location', 'Borough', "Handicap Accessible"]]
```

```python
df = df.fillna('No')
df
```

```python
def geocoding(address, borough):
    addresslist = address.strip().split()
    borough = borough.strip()
    gmapadd = ""
    for i in addresslist:
        gmapadd = gmapadd + "+" + i
    gmapadd = gmapadd + ",+" + borough + ",+New York"
    gmapadd = "https://maps.googleapis.com/maps/api/geocode/json?address=" + gmapadd + "&key=AIzaSyAi1163PQBVohNzZ1dfAAOiQ4hvwHkW1N0"
    
    
    r = requests.get(gmapadd)
    jsondata = json.loads(r.text)
    if len(jsondata['results']) == 0:
        coordinates = "Error"
    else:
        coordinates = jsondata['results'][0]['geometry']['location']
    return coordinates
```

```python
df['Coordinate'] = df[['Location', 'Borough']].apply(lambda x: geocoding(x['Location'], x['Borough']), axis = 1)
```

```python
df
```

```python
df['Coordinate2'] = df[df.Coordinate == "Error"].apply(lambda x: geocoding(x['Name'], x['Borough']), axis = 1)
```

```python
df=df.fillna('Error')
df
```

```python
df = df[(df.Coordinate != "Error") | (df.Coordinate2 != "Error")]
```

```python
def coordinate(coor1, coor2):
    if coor1 == "Error":
        coor = coor2
    else:
        coor = coor1
    return coor
```

```python
df['Coordinates'] = df.apply(lambda x: coordinate(x['Coordinate'], x['Coordinate2']), axis = 1)
```

```python
df
```

```python
locations=[]
for row, index in df.iterrows():
    instance = []
    address = index[1]+", "+index[2]
    instance.append(index[0])
    instance.append(address)
    instance.append(index[6]['lat'])
    instance.append(index[6]['lng'])
    locations.append(instance)
```
