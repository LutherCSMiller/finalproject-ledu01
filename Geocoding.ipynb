```python
import pandas as pd
import numpy as np
import json
import requests
%matplotlib inline
import matplotlib.pyplot as plt
df = pd.read_csv("Airplane_Crashes_and_Fatalities_Since_1908.csv")
```

```python
df = df[['Date', 'Time', 'Location', 'Operator', 'Aboard', 'Fatalities', 'Ground']]
df['Decade'] = df['Date'].apply(lambda x: ((int(str(x)[-4:])//10)*10))
df
```

```python
locations = {}
for i in range(1900, 2010, 10):
    locations[i] = []
    for j in df[df.Decade == i]['Location']:
        locations[i].append(j)
```

```python
coordinates = {}
for t in range(1900, 2010, 10):
    coordinates[t] = []
    for u in locations[t]:
        address = "https://maps.googleapis.com/maps/api/geocode/json?address="+str(u)+"key=AIzaSyAXtNIJroOwV-QzuC-vxRCP9wvNgv4tMzk"
        r = requests.get(address)
        jsondata = json.loads(r.text)
        if len(jsondata['results']) == 0:
            coordinates[t].append("Error")
        else:
            coordinates[t].append(jsondata['results'][0]['geometry']['location'])
```
